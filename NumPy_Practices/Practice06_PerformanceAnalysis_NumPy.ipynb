# Timing, Vectorization, Computational Complexity, and Memory Efficiency

```python
import numpy as np
import time
```

```python
# Timing Loops vs Vectorization

N = 1_000_000
x = np.random.rand(N)

# Loop
start = time.time()
y = np.zeros(N)
for i in range(N):
    y[i] = x[i] *2

end = time.time()
print("Loop time: ", end - start)

# Vectorized
start= time.time()

y = x * 2 

end = time.time()
print("Vectorized time: ", end - start)
```

```python
# Optimize a Slow Operation

# Let's compute the pairwise differences. Dij = (xi - xj)**2

N = 2000
x = np.random.rand(N)

# Loop
start = time.time()
D = np.zeros((N,N))
for i in range(N):
    for j in range(N):
        D[i,j] = (x[i] - x[j])**2 
# If N doubles, work becomes 4× larger.
# This is called O(N²) (quadratic growth).
end = time.time()
print("Loop time: ", end - start)

# Vectorized version using broadcasting:
start = time.time()
D = (x[:, None] - x[None, :])**2 # Still O(N²) mathematically, but implemented in compiled C.
# Note: memory complexity is also O(N²).
end = time.time()
print("Broadcasting time: ", end - start)

# Note: Temporary arrays created during vectorization increase memory usage.
```

Time complexity: describes how runtime grows with input size.

```python
# Memory Footprint

# memory = number-of-elements × bytes-per-element
start = time.time()
x = np.random.rand(1000,1000) # float64 ≈ 15 decimal digits
end = time.time()
print(x.nbytes, " in ", end - start) # (~8 MB)

start = time.time()
x32 = x.astype(np.float32) # float32 ≈ 7 decimal digits
end = time.time()
print(x32.nbytes, " in ", end - start) # (~4 MB)

# Note: memory affects time.
```

Note:
- Reaction time data is usually safe with float32.
- High-precision numerical optimization may require float64.
