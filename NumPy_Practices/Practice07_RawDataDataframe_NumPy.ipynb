# Data Loading, Validation, and Missing-Value Handling

```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
# Loading data
file_path = Path(
    r"F:\DocHob\Education\Programing\Python_Language\PythonProjects\Raw_Data\Exprement01_ManualMiniData_Raw.txt"
)

data = np.genfromtxt(
    file_path,
    delimiter=",",
    skip_header=1, # Imprtant. Do not forget.
    dtype=float
)
```

Note: do not use loadtxt when the file contains missing entries.

```python
# Validate Shape, dtype, Missing Values

print("Shape:", data.shape)
print("Dtype:", data.dtype)
print("Missing values:", np.isnan(data).sum())

rt = data[:,1]
print("RT min:", np.nanmin(rt))
print("RT max:", np.nanmax(rt))
# Ensure no negative RT, and no implausible extreme values.
```

```python
# Panda

columns = ["trial", "rt", "accuracy"]
df = pd.DataFrame(data, columns=columns)
print(df)
```

Note:
1. np.genfromtxt handles missing values; np.loadtxt does not.
2. Panda can handle missing values by itself.
