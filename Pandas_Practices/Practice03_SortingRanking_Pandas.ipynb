```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(1)
```

```python
df = pd.DataFrame({
    "Subject": np.repeat(["S1", "S2", "S3"], 5),
    "Trial": list(range(1, 6)) * 3,
    "RT": np.random.normal(loc=500, scale=50, size=15)
})
```

> sort_values(): sort by column valuse.

```python
# Basic syntax:
print(df.sort_values(by="RT").head())

# Descending order:
print(df.sort_values(by="RT", ascending=False).head())

# Multiple columns:
print(df.sort_values(by=["Subject", "RT"]).head())
```

```python
# Note: sorting does not modify the DataFrame unless reassigned. 

print(df.head())

df_sorted = df.sort_values(by="RT")
print(df_sorted.head())
```

> sort_index(): sort by row index rather than column values.

```python
print(df_sorted.sort_index().head())
# Note: useful after shuffling, concatenation, or hierarchical indexing.
```

> rank(): assigns ranks to values.

By default:
   - Smallest value gets rank 1.
   - Ties receive average rank.

Ranking methods:
   - method="average" (default)
   - method="min"
   - method="max"
   - method="first"
   - method="dense"

Note: ranking is scale-independent: it ignores magnitude differences.

```python
# Descending rank (fastest RT = highest rank):

df["RT_rank_desc"] = df["RT"].rank(ascending=False)

print(df.head())
```

```python
# Let's sort subjects by mean of RT.

subject_mean_rt = (
    df.groupby("Subject")["RT"] # groupby() partitions data. We will learn more.
    .mean()
    .sort_values()
)

print(subject_mean_rt)
```

```python
# Let's rank trials within each subject (fastest trial = rank 1).

df["RT_rank_within_subject"] = (
    df.groupby("Subject")["RT"]
      .rank()
)

print(df.head())
```
