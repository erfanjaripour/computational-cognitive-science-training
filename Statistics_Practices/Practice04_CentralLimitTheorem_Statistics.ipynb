```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(1)
```

```python
# Creating a skewed population.

population = np.random.exponential(scale=1, size=100000)

# Testing how skewed the popilation is.

print(pd.Series(population).skew())

# Visualizing the population.

plt.hist(population, bins=50)
plt.title("Population Distribution")

plt.show()
```

```python
# Population statistics.

s = pd.Series(population)

print("Mean:", s.mean())
print("Median:", s.median())
print("Skewness:", s.skew())
```

```python
# Creating repeated samples.

sample_means = []

for _ in range(5000):

    sample = np.random.choice(population, size=30)

    sample_means.append(
        sample.mean()
    )

# Visualizing mean distribution.

plt.hist(sample_means, bins=40)
plt.title("Sampling Distribution of Means")

plt.show()
```

```python
# Comparing standard deviations between sample means and population.

print(np.std(population))

print(np.std(sample_means))
```

Note: sampling is more normal and its SD is much smaller.
