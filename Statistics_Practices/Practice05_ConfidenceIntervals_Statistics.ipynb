```python
import numpy as np
from scipy import stats
```

```python
np.random.seed(1)
```

```python
# Computing CI manually.

data = np.array([95, 102, 100, 105, 98, 101, 97, 103, 99, 104])

# Computing mean.

mean = np.mean(data)

print("Mean:", mean)

# Computing sample standard deviation.

sd = np.std(data, ddof=1)

print("SD:", sd)

# Creating sample size based on data length.

n = len(data)

print("Sample size:", n)

# Computing standard error.

se = sd / np.sqrt(n)

print("SE:", se)

# Calculating margin of error.

me = 1.96 * se

print("Margin of error:", me)

# Calculating confidence interval.

lower = mean - me
upper = mean + me

print("Confidence interval is between", lower,"and", upper)
```

```python
# Computing CI with sciPy.

result = stats.t.interval(
    confidence=0.95,
    df=len(data)-1,
    loc=np.mean(data),
    scale=stats.sem(data)
)

print(result)
```

```python
# Showing the effect of sample sizes.

# Creating two different size samples.

small_sample = np.random.normal(loc=100, scale=15, size=20)

large_sample = np.random.normal(loc=100, scale=15, size=500)

# Computing CI for each one.

small_sample_CI = stats.t.interval(
    confidence=0.95,
    df=len(small_sample)-1,
    loc=np.mean(small_sample),
    scale=stats.sem(small_sample)
)

large_sample_CI = stats.t.interval(
    confidence=0.95,
    df=len(large_sample)-1,
    loc=np.mean(large_sample),
    scale=stats.sem(large_sample)
)

print("CI of the small sample:", small_sample_CI)

print("CI of the large sample:", large_sample_CI)
```

Note: notice that large sample has narrower CI.
