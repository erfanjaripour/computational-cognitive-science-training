```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
np.random.seed(1)
```

```python
# Computing Cohen's d manually.

# Creating datasets.

group1 = np.array([90,95,100,105,110])
group2 = np.array([100,105,110,115,120])

# Calculating means.

print("Group 1 mean:", group1.mean())
print("Group 2 mean:", group2.mean())

# Calculating standard deviations.

print("Group 1 standard deviation:",
      np.std(group1, ddof=1))
print("Group 1 standard deviation:",
      np.std(group1, ddof=1))

# Pooled SD.

pooled_sd = np.sqrt(
    (
        (len(group1)-1)*np.var(group1, ddof=1)
        +
        (len(group2)-1)*np.var(group2, ddof=1)
    )
    /
    (len(group1)+len(group2)-2)
)

# Cohen's d.

d = (
    group2.mean() -
    group1.mean()
) / pooled_sd

print("Cohen's d:", d)
```

```python
# Showing same p-values sometimes have different effect sizes.

# Dataset A: tiny effect, huge sample.

group_A1 = np.random.normal(loc=100, scale=10, size=10000)
group_A2 = np.random.normal(loc=101, scale=10, size=10000) # Very small differences between the samples.

# t-testing.

print(stats.ttest_ind(group_A1, group_A2))

# Calculating d.

mean_diff = np.mean(group_A2) - np.mean(group_A1)

pooled_sd = np.sqrt(
    (
        np.var(group_A1, ddof=1)
        +
        np.var(group_A2, ddof=1)
    ) / 2
)

d = mean_diff / pooled_sd

print("Cohen's d for dataset A:", d)

# Dataset B: large effect, smaller sample.

group_B1 = np.random.normal(loc=100, scale=10, size=100)
group_B2 = np.random.normal(loc=105, scale=10, size=100)

# t-testing.

print("\n", stats.ttest_ind(group_B1, group_B2))

# Calculating d.

mean_diff = np.mean(group_B2) - np.mean(group_B1)

pooled_sd = np.sqrt(
    (
        np.var(group_B1, ddof=1)
        +
        np.var(group_B2, ddof=1)
    ) / 2
)

d = mean_diff / pooled_sd

print("Cohen's d for dataset B:", d)
```
