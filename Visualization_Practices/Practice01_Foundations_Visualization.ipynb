```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creating a figure with plt.figure()
plt.figure(figsize=(8, 4))

x = [1, 2, 3, 4, 5]
y = [320, 310, 295, 290, 285]

plt.plot(x, y)

plt.title("Reaction Time Across Trials")
plt.xlabel("Trial Number")
plt.ylabel("Reaction Time (ms)")

plt.show()
```

Note: plt.plot() draws data onto axes.
> plt.plot(x_values, y_values)

```python
trials = [1, 2, 3, 4, 5]
reaction_time = [350, 340, 325, 315, 300]

plt.figure(figsize=(8, 4))

plt.plot(
    trials,
    reaction_time,
    marker='o',
    linestyle='--',
    label='Participant A'
)

plt.title("Reaction Time Improvement")
plt.xlabel("Trial")
plt.ylabel("Reaction Time (ms)")
plt.legend()

plt.show()
```

Note: It is better to use fig, ax = plt.subplots(); instead of using global plotting commands.

#### Pros of Object-oriented plotting:
- scales better for complex figures,
- supports multiple subplots,
- improves reproducibility,
- avoids state confusion,
- is standard in scientific Python ecosystems.

```python
fig, ax = plt.subplots(figsize=(8, 4)) # fig: entire figure, ax: axes object.

trials = np.array([1, 2, 3, 4, 5, 6])
reaction_time = np.array([410, 390, 370, 365, 350, 340])

ax.plot(
    trials,
    reaction_time,
    marker='o',
    label='Experimental Group'
)

ax.set_title("Reaction Time Across Learning Trials")
ax.set_xlabel("Trial")
ax.set_ylabel("Reaction Time (ms)")
ax.legend()

plt.show()
```

```python
# Let's create a mini dataset and create a plot of it.

trials = np.arange(1, 11)
reaction_times = np.array([420, 405, 398, 390, 382, 375, 370, 368, 365, 360])

fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(
    trials,
    reaction_times,
    marker='o',
    linewidth=2,
    label='Participant 1'
)

ax.set_title("Reaction Time Reduction Across Trials")
ax.set_xlabel("Trial Number")
ax.set_ylabel("Reaction Time (ms)")

ax.legend()

plt.show()
```

### Exploratory vs Publication Plots
#### Exploratory Plot Characteristics

Exploratory figures often include:
- dense annotations
- multiple variables
- diagnostic overlays
- raw datapoints
- temporary formatting

Example goals:
- detect outliers
- inspect variance
- identify artifacts

#### Publication Plot Characteristics
Publication figures prioritize:
- clarity
- readability
- interpretability
- minimal clutter
- accessibility

They often remove unnecessary detail.

```python
# Exploratory Example

fig, ax = plt.subplots(figsize=(10, 5))

noise = np.random.normal(0, 5, size=10)

experimental_rt = reaction_times + noise

ax.plot(trials, reaction_times, marker='o', label='Observed')
ax.plot(trials, experimental_rt, marker='x', label='Noisy Signal')

ax.grid(True)

ax.set_title("Exploratory Inspection of Reaction Time Signals")
ax.set_xlabel("Trial")
ax.set_ylabel("RT (ms)")
ax.legend()

plt.show()
```

```python
# Publication Example

fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(
    trials,
    reaction_times,
    linewidth=2
)

ax.set_title("Reaction Time Decreases Across Repeated Trials")
ax.set_xlabel("Trial")
ax.set_ylabel("Reaction Time (ms)")

plt.tight_layout()
plt.show()
```

#### Characteristics of Misleading Figures
- Truncated axes > Exaggerates differences
- Missing labels > Prevents interpretation
- Excessive smoothing > Hides variability
- Distorted aspect ratios > Alters trend perception
- Overlapping colors > Reduces readability
- No uncertainty intervals > Overstates certainty
- Cherry-picked ranges > Biases interpretation
- Decorative clutter > Distracts from data

#### Bad Plot 1: Truncated Y-axisc

```python
fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(trials, reaction_times, linewidth=3)

ax.set_ylim(350, 425)

ax.set_title("BAD EXAMPLE: Exaggerated Improvement")
ax.set_xlabel("Trial")
ax.set_ylabel("Reaction Time (ms)")

plt.show()
```

Problem:
- The y-axis begins at 350 instead of 0.
- Improvement appears visually dramatic.
- Magnitude perception becomes distorted.
- Effect size may appear stronger than reality.

#### Bad Plot 2: Missing Labels

```python
plt.figure(figsize=(8, 4))
plt.plot(trials, reaction_times)
plt.title("Data")
plt.show()
```

Problems:
- No x-axis label.
- No y-axis label.
- Ambiguous title.
- Impossible to interpret context.
- Figure becomes non-reproducible,
- Interpretation depends on external explanation.

#### Bad Plot 3: Visual Clutter

```python
fig, ax = plt.subplots(figsize=(12, 6))

for i in range(10):
    random_noise = np.random.normal(0, 10, size=10)
    ax.plot(trials, reaction_times + random_noise)

ax.grid(True)

ax.set_title("BAD EXAMPLE: Excessive Overplotting")

plt.show()
```

Problems:
- Excessive overlapping signals,
- No legend,
- No grouping logic,
- High cognitive load.
- Difficult pattern extraction,
- Poor interpretability,
- Obscures meaningful structure.

##### General Principles:
- Label everything clearly.
- Avoid unnecessary decoration.
- Preserve scale integrity.
- Use readable font sizes.
- Minimize cognitive load.
- Show uncertainty when relevant.
- Match visualization to scientific question.
- Prefer reproducibility over aesthetics.

#### Marker Control

Markers identify the actual observed data points.
- Without markers: viewers may over-trust the continuity.
- With markers: observed measurements become explicit.

```python
trials = [1, 2, 3, 4, 5]
scores = [10, 15, 18, 30, 28]

plt.plot(trials, scores, marker='o')

plt.xlabel("Trial")
plt.ylabel("Score")
plt.title("Scores Across Trials")

plt.show()
```

```python

plt.plot(
    trials,
    scores,
    marker='o',
    markersize=10,
    markeredgewidth=2
)
```

```python
# Multi-Line Plotting

trials = [1, 2, 3, 4, 5]

subject_A = [50, 55, 60, 68, 72]
subject_B = [48, 52, 58, 61, 66]

plt.plot(trials, subject_A, marker='o', label='Subject A')
plt.plot(trials, subject_B, marker='s', label='Subject B')

plt.xlabel("Trial")
plt.ylabel("Performance")
plt.title("Performance Across Trials")

plt.legend()

plt.show()
```

Axis Scaling: axis scaling fundamentally changes interpretation.

Linear Scale: equal visual distances represent equal numerical changes.

```python
# Linear scale

x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]

plt.plot(x, y, marker='o')
plt.title("Linear Scale")
plt.show()
```

Logarithmic Scale: equal visual distances represent multiplicative change.

```python
# Logarithmic scale

x = [1, 2, 3, 4, 5]
y = [1, 10, 100, 1000, 10000]

plt.plot(x, y, marker='o')
plt.yscale('log')
plt.title("Logarithmic Y-Axis")
plt.show() # exponential growth becomes approximately linear.
```

#### Compare Linear vs Logarithmic Scales

```python
# Dataset

days = [1, 2, 3, 4, 5]
bacteria = [10, 100, 1000, 10000, 100000]
```

```python
# Linear scale

days = [1, 2, 3, 4, 5]
bacteria = [10, 100, 1000, 10000, 100000]

plt.plot(days, bacteria, marker='o')

plt.xlabel("Day")
plt.ylabel("Population")
plt.title("Linear Scale")

plt.show()
```

```python
# Logarithmic scale

days = [1, 2, 3, 4, 5]
bacteria = [10, 100, 1000, 10000, 100000]

plt.plot(days, bacteria, marker='o')

plt.yscale('log')

plt.xlabel("Day")
plt.ylabel("Population (log scale)")
plt.title("Logarithmic Scale")

plt.show()
```

Note: Linear scale emphasizes absolute growth. Log scale emphasizes proportional growth. Neither is “correct.” They answer different questions.
