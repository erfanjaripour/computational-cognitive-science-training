```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
np.random.seed(1)
```

```python
hours_studied = [1, 2, 3, 4, 5]
exam_score = [52, 57, 63, 70, 78]

plt.scatter(hours_studied, exam_score)

plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.title("Study Time vs Exam Score")

plt.show()
```

#### Difference Between Line and Scatter Plots 
- Line plots implies continuity. Its main purpose is to show trends over ordered sequence.
- Scatter plot does not implies continuity. Its main purpose is to show relationship between variables.

Note: scatter plots are safer when observations are independent.

### Correlation

```python
# Simulating positive correlation.

x = np.random.normal(loc=50, scale=10, size=100)

noise = np.random.normal(loc=0, scale=5, size=100)

y = 2 * x + noise

plt.scatter(x, y)

plt.xlabel("X Variable")
plt.ylabel("Y Variable")
plt.title("Simulated Correlated Variables")

plt.show()
```

```python
# Creating a strong correlation plot.

x = np.random.normal(loc=0, scale=1, size=100)
y = 3*x + np.random.normal(loc=0, scale=0.5, size=100)

plt.scatter(x, y)
plt.title("Strong Correlation")

plt.show()
```

```python
# Creating a weak correlation plot.

x = np.random.normal(loc=0, scale=1, size=100)
y = 3*x + np.random.normal(loc=0, scale=4, size=100)

plt.scatter(x, y)
plt.title("Weak Correlation")

plt.show()
```

#### Fake Strong Correlation

```python
# Creating a fake strong correlation.

x = np.random.normal(loc=0, scale=1, size=100)
y = 0.5*x + np.random.normal(loc=0, scale=1, size=100)

# Wide axes

plt.scatter(x, y)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.title("Wide Axes")

print(plt.show())

# Zoomed axes

plt.scatter(x, y)
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.title("Zoomed Axes")

print(plt.show())
```

#### Overplotting

```python
# Creating a overplotting example.

x = np.random.randint(0, 10, size=1000)
y = np.random.randint(0, 10, size=1000)

plt.scatter(x, y)
plt.title("Overplotting Example")

plt.show()
```

##### Alpha Transparency
Alpha transparency solves overplotting partially.
> alpha = 0  (fully transparent)

> alpha = 1  (fully opaque)

```python
# Using alpha transparency.

x = np.random.randint(0, 10, size=1000)
y = np.random.randint(0, 10, size=1000)

plt.scatter(x, y, alpha=0.2)
plt.title("Transparency Reduces Overplotting")

plt.show()
```

Notice that darker areas imply more observations.

##### Regression Overlays
Regression overlays estimate the average relationship between variables.

It automatically:
- Draws scatter points.
- Estimates linear regression.
- Plots regression line.
- Shows confidence interval.

```python
# Creating regression plot in seaborn.

x = np.random.normal(loc=0, scale=1, size=100)
y = 2*x + np.random.normal(loc=0, scale=1, size=100)

df = pd.DataFrame({'x': x, 'y': y})

sns.regplot(data=df, x='x', y='y') 

plt.title("Regression Overlay")

plt.show()
```

#### Histogram

```python
# Creating a histogram plot.

data = np.random.normal(loc=0, scale=1, size=1000)
plt.hist(data)
plt.title("Basic Histogram")

plt.show()
```

```python
# Creating a normal distributed histogram.

normal_data = np.random.normal(loc=50, scale=10, size=1000)

plt.hist(normal_data, bins=30)
plt.title("Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```

```python
# Creating a right-skewed distributed histogram.

skewed_data = np.random.exponential(scale=1, size=1000)

plt.hist(skewed_data, bins=30)
plt.title("Right-Skewed Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```

#### KDE (Kernel Density Estimation)

```python
# Creating a KDE plot.

data = np.random.normal(loc=0, scale=1, size=1000)

sns.kdeplot(data)

plt.title("Kernel Density Estimate")

plt.show()
```

```python
# Creating histogram and KDE together.

data = np.random.normal(loc=0, scale=1, size=1000)

sns.histplot(
    data,
    bins=30,
    kde=True
)

plt.title("Histogram with KDE")

plt.show()
```

#### Binning Problems
Binning choices can entirely change the appearance of a histogram.

```python
# An example of binning problem

data = np.random.normal(loc=0, scale=1, size=200)

# Too few bins

plt.hist(data, bins=5)
plt.title("Too Few Bins")

print(plt.show())

# Too many bin

plt.hist(data, bins=100)
plt.title("Too Many Bins")

print(plt.show())
```

```python
# Anothe example of binning problem

data = np.random.normal(loc=0, scale=1, size=500)

plt.figure(figsize=(15,4))

plt.subplot(1,3,1)
plt.hist(data, bins=5)
plt.title("5 Bins")

plt.subplot(1,3,2)
plt.hist(data, bins=20)
plt.title("20 Bins")

plt.subplot(1,3,3)
plt.hist(data, bins=80)
plt.title("80 Bins")

plt.show()
```

#### Outliers

```python
# An example of outliers

main_data = np.random.normal(0, 1, 500)
outliers = [8, 9, 10]
data = np.concatenate([main_data, outliers])

plt.hist(data, bins=40)
plt.title("Histogram with Outliers")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```

#### Boxplots

```python
# Creating a boxplot.

data = np.random.normal(loc=50, scale=10, size=300)

plt.boxplot(data)
plt.title("Basic Boxplot")

plt.show()
```

```python
# Creating boxplot with outliers.

main_data = np.random.normal(50, 5, 200)
outliers = [90, 95, 100]
data = np.concatenate([data, outliers])

plt.boxplot(data)
plt.title("Boxplot with Outliers")

plt.show()
```

Note: boxplot “outliers” are not necessarily errors, anomalies, or invalid observations. They are merely statistically distant observations.

Note: boxplots hide distribution shape. Very different datasets can produce identical boxplots.

#### Demonstration: Hidden Multimodality

```python
# Creating a unimodal dataset.
data_unimodal = np.random.normal(loc=50, scale=8, size=500)

# Creating a bimodal dataset
part1 = np.random.normal(loc=40, scale=4, size=250)
part2 = np.random.normal(loc=60, scale=4, size=250)
data_bimodal = np.concatenate([part1, part2])

# Boxplots of both
plt.boxplot([data_unimodal, data_bimodal])
plt.xticks([1,2], ['Unimodal', 'Bimodal'])
plt.title("Boxplots Hide Multimodality")

plt.show() # Bimodal dataset contains two populations, two peaks, and latent subgroup structure.
```

Note: Boxplot destroys local density information.

#### Violin Plots

```python
# Creating a violin plot.

data = np.random.normal(loc=0, scale=1, size=300)

sns.violinplot(y=data)
plt.title("Basic Violin Plot")

plt.show()
```

Note: unlike boxplots, violin plots can reveal skewness, multimodality, asymmetry, and density concentration. But violin plots inherit all KDE limitations.

```python
# Creating a violin plot to show it reveals bimodality.

part1 = np.random.normal(loc=-2, scale=0.5, size=300)
part2 = np.random.normal(loc=2, scale=0.5, size=300)

data = np.concatenate([part1, part2])

sns.violinplot(y=data)
plt.title("Violin Plot Reveals Bimodality")

plt.show()
```

```python
# Creatin a violin plot to show it has KDE smoothing bias

data = np.random.normal(loc=0, scale=1, size=80)

sns.violinplot(y=data)
plt.title("Small Sample Violin Plot")

plt.show()
```

##### Using Multiple Visualizations

```python
# creating a dataset for comparing two conditions in boxplot and violin plot.
control = np.random.normal(loc=50, scale=5, size=200)
treatment = np.random.normal(loc=58, scale=7, size=200)

# Boxplot comparsion
plt.boxplot([control, treatment])
plt.xticks([1,2], ['Control', 'Treatment'])
plt.title("Condition Comparison with Boxplots")

print(plt.show())

# Violin plot comparsion
sns.violinplot(data=[control, treatment])
plt.xticks([0,1], ['Control', 'Treatment'])
plt.title("Condition Comparison with Violins")

print(plt.show())
```

```python
# Creating dataset for comparing histogram, boxplot, and violin versions.
part1 = np.random.normal(40, 5, 300)
part2 = np.random.normal(70, 5, 300)
data = np.concatenate([part1, part2])

# Histogram version
plt.hist(data, bins=30)
plt.title("Histogram")

print(plt.show())

# Boxplot version
plt.boxplot(data)
plt.title("Boxplot")

print(plt.show())

# Violin plot version
sns.violinplot(y=data)
plt.title("Violin Plot")

print(plt.show())
```
