```python
import pandas as pd
import numpy as np
```

```python
# Series of Reaction Times

# 10 subject's reaction times in milliseconds
rt_series = pd.Series(
    [523, 610, 487, 590, 705, 450, 678, 512, 560, 630],
    name= "ReactionTime_ms"
) # A pandas Series is a one-dimensional labeled array.

# Compute statistics
mean_rt = rt_series.mean() # arithmetic average
median_rt = rt_series.median() # robust central tendency
std_rt = rt_series.std() # sample standard deviation (ddof=1 by default)

mean_rt, median_rt, std_rt
```

```python
# DataFrame

data = {
    "SubjectID": [f"S{i+1}" for i in range(10)],
    "Age": [22, 25, 19, 24, 30, 21, 27, 23, 26, 28],
    "RT": [523, 610, 487, 590, 705, 450, 678, 512, 560, 630]
} # A DataFrame is a two-dimensional labeled data structure

df = pd.DataFrame(data)

# Inspect structure
print("Head:\n", df.head(), "\n\n")
print("Tail:\n", df.tail(), "\n\n")
print("Info:\n", df.info(), "\n\n")
print("Describe:\n", df.describe(), "\n\n")
print("Dtypes:\n", df.dtypes)
```

```python
# Convert RT from seconds to milliseconds

df["RT"] = df["RT"] / 1000  # convert to seconds artificially

# Convert back to milliseconds:
df["RT_ms"] = df["RT"] * 1000
```

Note: Always create a new column when transforming units in a research notebook to preserve provenance.

Series and DataFrames are foundational. These structures:  
- Keep variable names and subject identifiers attached to the data.
- Allow fast, column-wise numerical operations without manual loops.
- Handle missing values in a structured and consistent way.
- Work directly with common statistical and machine learning libraries.
- Support fully script-based analyses that can be rerun and reproduced exactly

Note: Indexing and selection are not merely syntactic conveniences; they determine whether preprocessing steps are logically valid. Many analytic errors originate from incorrect subsetting rather than incorrect statistical modeling.

```python
# loc vs iloc

print(df.loc[0, "RT"]) # label-based (row label 0, column RT)
print(df.iloc[0, 2])  # first row, third column by position
```

```python
# Conditional Selection (Boolean Masking)

mask = df["Age"] > 25
print(df[mask])
# or
print(df[(df["Age"] > 25) & (df["RT"] < 600)])
```

  The mask must: 
1. Be the same length as the DataFrame.
2. Be enclosed in parentheses when combined with logical operators.

```python
# Select subjects with Age > 25

df[df["Age"] > 25] # Or df.loc[df["Age"] > 25]: this isolates older participants for subgroup analysis.
```

```python
# Filter trials where RT < 200 ms or RT > 2000 ms (between)

df[~df["RT"].between(200,2000)] # Or df[(df["RT"] < 200) | (df["RT"] > 2000)]
# The tilde ~ negates the condition.
```

```python
# Checks membership in a set (isin)

print(df[df["SubjectID"].isin(["S1", "S3", "S5"])])
```

```python
# Subset columns: only SubjectID and RT

df[["SubjectID", "RT"]] # Double brackets are required because you are passing a list of column names.
# df["SubjectID", "RT"] is incorrect, interpreted as tuple
```

Note:
- Always inspect the result with head().
- Confirm row counts before and after filtering.
- Explicitly report filtering criteria in markdown.
