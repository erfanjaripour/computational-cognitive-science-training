```python
def greet(names):
    for name in names:
        print("Hello", name)

names = ["a","b","c"]
greet(names)
```

```python
def factorial(x):
    if x==0:
        return 1
    else:
        return x*factorial(x-1)

factorial(5)
```

```python
def BMI(waight, hight):
    return waight/(hight*hight)
BMI(60,1.78)
```

```python
from math import sqrt, sin, factorial
def mathmath(x):
    print(sqrt(x), sin(x), factorial(x))

mathmath(5)
```

```python
def isprime(x):
    if x <= 1:
        return False
    else:
        n = x
        while n > 2:
            if x % (n-1) == 0:
                return False
            else: n = n-1
        return True

isprime(-2)
```
