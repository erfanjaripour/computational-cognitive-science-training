```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(1)
```

```python
# Creating repeated samples to show sampling variability.

sample_means = []

population = np.random.normal(loc=100, scale=15, size=100000)

for _ in range(1000):
    sample = np.random.choice(
        population,
        size=50
    )

    sample_means.append(
        sample.mean()
    )

sample_means = np.array(sample_means)

print(sample_means[:10])
```

```python
# Comparing sample mean with population mean.

print("Population mean:", population.mean())

print("Sample mean:", sample_means.mean())
```

Note: Individual samples vary. But the average of many sample means is very close to population mean.

```python
# Visualizing sampling variability.

plt.hist(sample_means, bins=30)

plt.axvline(
    population.mean(),
    linestyle='--'
)

plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.title("Sampling Distribution")

plt.show()
```
