```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
np.random.seed(1)
```

```python
# Using histogram to see distributions.

reaction_times = [420, 450, 470, 490, 500, 520, 540, 560, 580]

plt.hist(reaction_times, bins=5)

plt.xlabel("Reaction Time")
plt.ylabel("Frequency")

plt.show()
```

#### Normal Distribution

```python
# Simulating a normal data.

normal_data = np.random.normal(loc=100, scale=15, size=10000)

# Visualizing the normal data.

plt.hist(normal_data, bins=40)

plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Normal Distribution")

plt.show()
```

Note: The 68-95-99.7 Rule
- About 68% lie within mean -+ 1 SD.
- About 95% lie within mean -+ 2 SD.
- About 99.7% lie within mean -+ 3 SD.

#### Skewness

```python
# Simulating and visualizing positively skewed data.

skewed_data = np.random.exponential(scale=1, size=10000)

plt.hist(skewed_data, bins=40)
plt.title("Positively Skewed Distribution")

plt.show()
```

```python
# Computing skewness with pandas.

print(pd.Series(normal_data).skew())

print(pd.Series(skewed_data).skew()) # Output often 1 to 3
```

#### Kurtosis

```python
# Computing kurtosis with pandas.

print(pd.Series(normal_data).kurt()) # Tipically near 0 for normal distributions.
```

Note

Positive:
- Heavy tails
- More outliers

Negative:
- Light tails
- Fewer outliers

```python
# Comparing robustness to outliers.

outlier_data = np.append(normal_data, 1000)

print("Normal mean:", np.mean(normal_data),"\nOutlier mean:", np.mean(outlier_data))
print("\nNormal median:", np.median(normal_data),"\nOutlier median:", np.median(outlier_data))
print("\nNormal standard deviation:", np.std(normal_data),"\nOutlier standard deviation:", np.std(outlier_data))
print("\nNormal skewness:", pd.Series(normal_data).skew(),"\nOutlier skewness:", pd.Series(outlier_data).skew())
print("\nNormal kurtosis:", pd.Series(normal_data).kurt(),"\nOutlier kurtosis:", pd.Series(outlier_data).kurt())
```

```python
# Creating a sampling distribution.

population = np.random.normal(loc=100, scale=15, size=100000)

# Collecting sample means.

sample_means = []

for _ in range(1000):
    sample = np.random.choice(
        population,
        size=50
    )
    sample_means.append(
        sample.mean()
    )

# Viualization
plt.hist(
    sample_means,
    bins=30
)

print(plt.show())

print("Population standard deviation:", population.std(), "Sample standard deviation:", np.std(sample_means))
```
