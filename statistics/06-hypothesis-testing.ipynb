```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
np.random.seed(1)
```

```python
# Simulating a null distribution.

# Supposing H0: population mean is 80.

# Creating the population.

population = np.random.normal(loc=80, scale=10, size=100000)

# Creating many samples form it

sample_means = []

for _ in range(5000):

    sample = np.random.choice(population,size=30)

    sample_means.append(
        sample.mean()
    )

# Visualizing.

plt.hist(sample_means, bins=40)

plt.title("Null Distribution")

plt.show()

# Calculating p value.

sample_means = np.array(sample_means)

p_value = np.mean(sample_means >= 85)

print("P value:", p_value)
```

```python
# Simulating many studies.

significant = 0

for _ in range(1000):

    sample = np.random.normal(loc=80, scale=10, size=30)

    p = stats.ttest_1samp(sample, 80).pvalue

    if p < 0.05:
        significant += 1

print(significant / 1000)
```
