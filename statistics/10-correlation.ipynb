```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import (pearsonr, spearmanr)
```

```python
np.random.seed(1)
```

```python
# Creating a perfect linear correlation.

x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

r, p = pearsonr(x, y)

print("Pearson r:", r)

rho, p = spearmanr(x, y)

print("Spearman r:", rho)
```

```python
# Creating and measuring a strong correlation.

x = np.arange(100)
y = x + np.random.normal(loc=0, scale=5, size=100)

# Visualizing.

plt.scatter(x, y)

plt.xlabel("X")
plt.ylabel("Y")

plt.show()

# Computing Pearson.

r, p_p = pearsonr(x, y)

print("Pearson r:", r)
print("Pearson p:", p_p)

# Computing Spearman.

rho, s_p = spearmanr(x, y)

print("Spearman r:", rho)
print("Spearman p:", s_p)
```

```python
# # Creating and measuring a weak correlation.

x = np.random.normal(loc=0, scale=1, size=100)
y = np.random.normal(loc=0, scale=1, size=100)

# Visualizing.

plt.scatter(x, y)

plt.show()

# Computing Pearson.

r, p_p = pearsonr(x, y)

print("Pearson r:", r)
print("Pearson p:", p_p)

# Computing Spearman.

rho, s_p = spearmanr(x, y)

print("Spearman r:", rho)
print("Spearman p:", s_p)
```

```python
# Generating nonlinear data.

x = np.linspace(-10, 10, 200)
y = x**2

# Visualizing.

plt.scatter(x, y)

plt.show()

# Computing Pearson.

r, p_p = pearsonr(x, y)

print("Pearson r:", r)
print("Pearson p:", p_p)

# Computing Spearman.

rho, s_p = spearmanr(x, y)

print("Spearman r:", rho)
print("Spearman p:", s_p)
```
