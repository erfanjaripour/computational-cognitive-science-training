```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(1)
```

```python
# Creating a plotting function.

def plot_series(
    x,
    y,
    xlabel,
    ylabel,
    title
):

    fig, ax = plt.subplots(
        figsize=(4,3)
    )

    ax.plot(
        x,
        y,
        linewidth=2
    )

    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_title(title)

    return fig, ax
```

```python
# Useing it wherever.

x = np.arange(100)
rt = np.random.normal(loc=500, scale=50, size=100)

plot_series(
    x,
    rt,
    "Trial",
    "RT (ms)",
    "Reaction Time"
)

plt.show()
```

```python
# Building a figure template.

def create_figure():

    fig, ax = plt.subplots(

        figsize=(4,3)

    )

    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)

    return fig, ax

# Now every figure generally will look the same.
```
